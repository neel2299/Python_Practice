```python
#Question 7 activity questions from the week 
#MAKE A DICTIONARY WHERE VAL ARE SET OF FACTORS OF THE KEYS. AND THEN FIND THE PAIR(S) WITH THE MOST COMMON  FACTORS
dic={}
def get_fac(element):
    lis=[]
    for i in range(1,element+1):
        if element%i==0:
            lis.append(i)
    return lis
for i in range(1,101):
    dic[i]=set(get_fac(i))
maxx=[]
def common(x,y):
    var=0
    for i in x:
        if i in y:
            var+=1
    return var
ans=0
for i in range(1,101):
    for j in range(1, 101):
        temp=common(dic[i],dic[j])
        if i!=j and ans<temp:
            ans=temp
            maxx=[(i,j)]
        elif i!=j and ans==temp:
            maxx.append((i,j))
print(ans,maxx)
```

```python
def merge(dic1,dic2,pri):
    dic={}
    if pri=="first":
        dic={}
        for i in dic2.keys():
            dic[i]=dic2[i]
        for i in dic1.keys():
            dic[i]=dic1[i]
    if pri=="second":
        dic={}
        for i in dic1.keys():
            dic[i]=dic1[i]
        for i in dic2.keys():
            dic[i]=dic2[i]
    return dic
```
